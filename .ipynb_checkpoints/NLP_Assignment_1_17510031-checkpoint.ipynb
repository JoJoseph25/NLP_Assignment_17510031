```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from nltk.corpus import wordnet as wn
```

```python
def msplit(string, sep):
    docs = [string]
    for char in sep:
        words = []
        for substr in docs:
            words.extend(substr.split(char))
        docs = words
    return docs

def inverse_dict(d):
    invr_dict={}
    for key in d:
        check = invr_dict.get(d[key],0)
        if check==0:
            invr_dict[d[key]] = [key]
        else:
            invr_dict[d[key]].append(key)
    return invr_dict

def histogram(word_list):
    # word_list=list(s.split())
    word_dict= {}
    for word in word_list:
        word_dict[word]= word_dict.get(word,0)+1
    return word_dict

def word_freq_descending(L):
    d1={}
    for word in L:
        d1[word] = d1.get(word,0) + 1
    inverse_d1= inverse_dict(d1)
    sorted_freq=list(inverse_d1.keys())
    sorted_freq.sort()
    sorted_freq=sorted_freq[::-1]
    Final_list=[]
    for freq in sorted_freq:
        repeat= freq
        for word in inverse_d1[freq]:
            for i in range(repeat):
                Final_list.append(word)
    word_dict=histogram(Final_list)
    return word_dict

def count_vocab(list_words):
    N = []
    i=0
    V=[]
    v = set()
    for word in words:
        i=i+1
        if word not in V:
            v.add(word)
        V.append(len(v))
        N.append(i)
        #print("Computing Vocab "+ str(int(i/len(words)*100))+"%"+" complete")
    return N,V
```

```python
'''
Importing Tom Sawyer Data and segmenting each word into list and creating a word frequency dictonary
'''
file = open("Tom_Sawyer.txt", 'r')
doc=msplit(file.read(),(" ","\n",",","(",")","-","--","\"","[","]","!","?","_",".",";",":","”","“"))
file.close()
words=list(filter(lambda x: x != "", doc))
word_dict=word_freq_descending(words)
```

```python
'''
Calculating Types, Token and TTR 
'''
types=len(word_dict.keys())
token=len(words)
TTR=types/token

print("Types= "+ str(types))
print("Token= "+ str(token))
print("Type/Token Ratio= "+ str(TTR)+"\n")
print("TTR is very low, which indicates that text does not have a very rich vocabulary \
or in other words has relatively \nless lexical variation")
```

```python
'''
Making a data frame for easier calculation
'''
word_df=pd.DataFrame.from_dict(word_dict,orient="index")
word_df.columns=["Freq"]
word_df["Rank"]= [i for i in range(1,len(word_df)+1)]
word_df["f.r"]= word_df["Freq"]*word_df["Rank"]
word_df["prob_r"]= word_df["Freq"]/token
word_df["A"]=word_df["prob_r"]*word_df["Rank"]
```

```python
'''
Zipf's Law for log(word rank) vs log(frequency) 
'''
x=np.log(pd.Series(word_df["Rank"]).values)
y=np.log(pd.Series(word_df["Freq"]).values)

plt.plot(x,y, "r")
plt.xlabel("Log(Rank)")
plt.ylabel("Log(Frequency)")
plt.title("Zipf's Law to Log(Word Rank) and their corresponding Log(Frequency)")
plt.show()

mean_A=np.mean(pd.Series(word_df["A"].values))
print("A is " +str(mean_A)+ ". \nFrom the plot it can be seen it is almost a straight line, except \
at the ends where the frequency \nof words are very less or very high. Mostly this text document obeys Zipf's Law")
```

```python
'''
Another way to plot Zipf's Law

This method is considered for comparison as several context words do not have meaning and using \
leads to inf values. To avoid such cases we will visualise data using similar graph.
'''
x=pd.Series(word_df["Rank"]).values
y=pd.Series(word_df["Freq"]).values

plt.plot(x,y, "r")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.title("Zipf's Law to Log(Word Rank) and their corresponding Log(Frequency)")
plt.show()
```

```python
'''
Calculating Number of Meaning for each word
'''
meanings=[]
for word in list(word_df.index.values):
    diff_types= wn.synsets(word)
    meanings.append(len(diff_types)+1)
word_df["no_meanings"]= meanings
```

```python
'''
Ploting Curve for Number of Meaning vs the square root of Word Rank
'''
x=np.sqrt(pd.Series(word_df["Rank"]).values)
y=pd.Series(word_df["no_meanings"]).values

def func(r,k):
    return k/r
popt, pcov = curve_fit(func, x, y)

plt.scatter(x,y, c="y",s=2, label="Text Data")
# plt.plot(x,y)
plt.plot(x, func(x, *popt), "r--",label="Curve Fit")
plt.xlabel("Square Root(1/Rank)")
plt.ylabel("No. of Word Meaning")
plt.title("Zipf's Law to Number of Meaning and square root of their corresponding Word Rank")
plt.legend()
plt.show()

first=np.mean(word_df["no_meanings"][:1000])
last=np.mean(word_df["no_meanings"][-1:len(word_df)-1000:-1])

print("The plot is similar to one we seen above for Zipf's Law but the curve is not as step, this could be as \
many function words do not have any associcated meaning but have much higher rank or common words with many meanings\
are frequently used in various conext and the words with lower rank are less in frequency and have only few meaning \
associated with them")
print("\nFor example: \n For first 1000 words mean number of meanings: "+str(first)\
      +"\n For last 1000 words mean number of meanings: "+str(last))
```

```python
'''
Ploting Curve for Length of Word vs Frequency
'''
word_len=[len(word) for word in word_dict.keys()]
word_df["word_len"]= word_len

x=pd.Series(word_df["Freq"]).values
y=pd.Series(word_df["word_len"]).values

def func(f,k):
    return k/f
popt, pcov = curve_fit(func, x, y)

plt.scatter(x,y, c="y",s=3,label="Text Data")
plt.plot(x, func(x, *popt), "r--",label="Curve Fit")
plt.xlabel("Frequency")
plt.ylabel("Word length")
plt.title("Zipf Law to Word length and their corresponding Frequency")
plt.legend()
plt.show()
print("The plot is similar to one we seen above for Zipf's Law but the peak is not as high as the earlier plot, this could be as \
many frequent words have smaller word length compared to longer words that have very few frequency")

max_length=np.amax(pd.Series(word_df["word_len"]).values)
long_word=word_df[word_df["word_len"]==18].index.values
long_freq=int(word_df[word_df["word_len"]==18].Freq.values)

min_word=word_df[word_df["word_len"]==1].index.values
min_freq=np.sum(word_df[word_df["word_len"]==1].Freq.values)
print("\nFor example: \n The longest length word("+str(long_word[0])+") of length "\
      + str(max_length) +" occurs "+ str(long_freq)+" times."
      +"\n The shortes length words("+str(min_word)+") of length 1" +" occur "\
      + str(min_freq)+" times.")
```
